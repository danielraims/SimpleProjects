```python
import pandas as pd
import statsmodels
import matplotlib.pyplot as plt
import numpy as np
```

# 1. Introduction and data analysis

We will use the "microsoft.csv" data downloaded from www.kaggle.com. It is a time series with 2192 observations, from 01-04-2015 to 31-03-2021. Below, we will see the graph and the prices of the stock between 01-01-2015 and 05-04-2015.

```python

microsoft =pd.read_csv('data/Microsoft_Stock.csv')
microsoft['Date'] = pd.to_datetime(pd.to_datetime(microsoft['Date']).dt.date)
microsoft = microsoft.set_index('Date').asfreq('D')
microsoft = microsoft.fillna(method = 'ffill')['Close']
display(microsoft.head())
microsoft.plot()
```

From the graph, we can see there's a trend, no clear seasonal pattern, but the variance changes over time, which is a real problem when dealing with time series. To fix this, we often use Box-Cox transformation or the logarithmic function. We'll go with the second option, and here's the resulting graph:

```python
log_microsoft = np.log(microsoft)
log_microsoft.plot()
```

From the last graph, we can confirm that the variance problem is resolved. It's often practical and more reasonable to split the data into two parts: training data and validation data. The reason is simple – we train the model on the training data and then validate it on the validation data it has never seen before. Our goal is to accurately predict future data that we haven't encountered yet. We'll take the first 1800 values as the training data and the rest to validate our model.

```python
size_data_train = 1800
data_train = pd.Series(log_microsoft[:1800])
data_test = pd.Series(log_microsoft[1800:])
data_train.plot(figsize=(10,3))
data_test.plot()
```

##### Trend estimation

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
X= np.arange(1800).reshape(-1,1)
y = data_train.values
linModel = linReg.fit(X,y)
pd.Series(linModel.predict(np.arange(len(data_train)).reshape(-1,1)), index= data_train.index).plot()
data_train.plot()
```

# 2.Stationarity

By definition, a series is (weakly) stationary if its mean and variance do not change over time. Stationarity is important because it's easier to study and estimate the parameters of a series that doesn't change much over time. This leads to better predictions. We've observed that the series has an upward trend, so it's not yet stationary.

Often, to make a non-stationary series stationary, we can use differencing. We'll choose:

$$X_t = X_t - X_{t-1} = \nabla X_t$$

##### Stationary test (test de Dickey-Fuller)

```python
from statsmodels.tsa.stattools import adfuller
print('p-value: {}'.format(adfuller(data_train)[1]))
```

This suggests that our series is not yet stationary. Now, we will difference the series and observe the result.

```python
fig, axes = plt.subplots(nrows = 1 ,ncols = 2, figsize = (12,4))
data_train_stationary = data_train.diff()
data_train_stationary.plot(ax= axes[1])
data_train.plot(ax = axes[0])
```

```python
print('p-value: {}'.format(adfuller(data_train_stationary.dropna())[1]))
```

We observe that the series is stationary with a p-value lower than 0.05 after just one differencing. This will be very useful later on for the integrated parameter of the ARIMA model.

# 3- Error Measurement and Model Quality

### AIC (Akaike Information Criterion)

The Akaike Information Criterion is an estimator of prediction error and can be used for comparing multiple models. Given a collection of models trained on the same data, AIC estimates the quality of each model. Thus, AIC can be used to select the optimal model. To calculate AIC, the following formula can be used: $$AIC = -2(\text{log likelihood}) + 2K$$ where K is the number of parameters of the model, and we use the log likelihood of the model. The smaller the AIC, the better the model. We will use this property to build the auto-ARIMA algorithm.

### RMSE (Root Mean Squared Error)

The Mean Squared Error (MSE) is the most popular prediction measure in statistics and machine learning for quantitative variables. It is an estimator that measures the average of the squared errors of predictions, which means the average of the differences between the actual values and the values estimated by the model squared. RMSE is the square root of MSE. The formula for MSE is as follows: For ($y_1,y_2,...,y_n$), n observations and $\hat{y}_1,\hat{y}_2,...,\hat{y}_n$ the estimates given by the model: $$MSE = \sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$. The smaller this value is, the better the model.

### MAPE (Mean Absolute Percentage Error)

While RMSE is a good error measurement tool, sometimes it's more reasonable to have a prediction score between 0 and 1. If we have an MSE of 342.71, we can't determine if the model is good or bad. However, if the score is between 0 and 1, for example, if the score is 0, then we can confidently say the model is either bad or good. The MAPE value ranges from 0 to 1 and is more popular in time series than in machine learning. The formula to use is: $$MAPE = \frac{1}{n} \sum_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{y_i}$$ where ($y_1,y_2,...,y_n$) are n observations and $\hat{y}_1,\hat{y}_2,...,\hat{y}_n$ are the estimates given by the model. The closer this value is to zero (0), the better the model. Sometimes, 1 - MAPE is used.


# 4. ARIMA and Model Parameter Estimation

## ARIMA (Autoregressive Integrated Moving Average)

ARIMA is a statistical model used for time series analysis with the goal of explaining the series or predicting future values. 

- **Autoregressive (AR) Model:** Predicts future values based on past values. An AR(p) model with \( p > 0 \) is represented by:
  $$ X_t = \phi_0 + \phi_1 X_{t-1} + \phi_2 X_{t-2} + ... + \phi_p X_{t-p} + \epsilon_t $$
  
- **Moving Average (MA) Model:** Models the error term as a linear combination of past error terms. An MA(q) model with \( q > 0 \) is represented by:
  $$ X_t = \theta_0 + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + ... + \theta_q \epsilon_{t-q} +\epsilon_t $$

- **ARMA Model:** Combines autoregressive and moving average components. An ARMA(p,q) model is represented by:
  $$ X_t = \theta_0 + \theta_1 X_{t-1} + \theta_2 X_{t-2} + ... + \theta_q X_{t-q} + \phi_1 X_{t-1} + \phi_2 X_{t-2} + ... + \phi_p X_{t-p} $$

- **ARIMA Model:** A more general model that also includes differencing to make the series stationary. An ARIMA(p,d,q) model has a parameter \( d \) representing the number of differences needed to achieve stationarity. The typical value of \( d \) is 1 or 2.


### Estimation of ARIMA Parameters using ACF and PACF

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize =(10,4))
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(data_train_stationary.dropna(),ax = axes[1]);
plot_pacf(data_train_stationary.dropna(), ax= axes[0], method ='yw');
```

The first thing we notice is that the ACF and PACF plots are very similar. Both the ACF and PACF show significant spikes at lag 2. Therefore, we will choose \( p = 2 \) for the autoregressive component and \( q = 2 \) for the moving average component.


### Building the model

```python
from pmdarima.arima.arima import ARIMA
model = ARIMA(order=(2,1,2))
model_fit = model.fit(data_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error

def metrics(true_data, predicted, model):
    print('        root mean squared error: ',mean_squared_error(true_data,predicted)**(1/2))
    print('                            aic: ',model_fit.aic())
    print(' mean_absolute_percentage_error: ',mean_absolute_percentage_error(true_data,predicted))
predicted = model_fit.predict_in_sample()
predicted = pd.Series(predicted, index =data_train.index)
metrics(np.exp(data_train), np.exp(predicted), model_fit)
```

First, we note that we are still working with the **training data**. We observe that approximately 99.3% (1 - MAPE) of the series is explained by the model, which is a very good score. However, we will see in the forecasting section if the model generalizes well to data it has not seen.


```python
model_fit.summary()
```

**Interpretation**:

We observe that some coefficients of the model are not significant, such as the coefficient of \( X_{t-2} \) and \( \epsilon_{t-2} \). This suggests that there might be room for improvement in the model. Additionally, the p-value of the Ljung-Box test or Portmanteau test is significant (p-value = 0.96 > 0.05), indicating that the residuals are independent and uncorrelated. 

The Jarque-Bera test result is somewhat disappointing, but it is logical since we observed that the data itself does not follow a normal distribution. However, we can see from the graph above that the ACF of the residuals is no longer correlated.



```python
plot_acf(model_fit.resid());
```

## Auto ARIMA -Recherches des paramètres optimaux


```python
from pmdarima import auto_arima
import warnings
warnings.filterwarnings(action='ignore')

best_model = auto_arima(data_train,start_p=0,start_d=0,start_q=0,max_p=5,max_q=5,max_d=2,)
best_model
```

We observe that the optimal parameters for the model are **p = 0, d = 1, and q = 4**. We will use these parameters for forecasting.


```python
best_model.summary ()
```

There isn't much difference between the last model and this one. The AIC is slightly smaller than that of the last model.


# Prevision

Now, let's see if our model generalizes well on the data it has never seen before. This is the sole purpose of this project. It's important to remember that we transformed the series using the logarithmic function. We will now use the exponential function to revert the series back to its original values.


```python
predicted = model.predict(len(data_test))
predicted = pd.Series(predicted, index =data_test.index)
np.exp(predicted).plot()
np.exp(data_train).plot()
np.exp(data_test).plot()
```

```python
metrics(np.exp(data_test), np.exp(predicted),best_model)
```

We can see that **95%** of the data is explained by the model, slightly lower than that of the training data. However, it's still a good score. Next, we will construct confidence intervals at a 0.05 confidence level for this model.


```python
predicted, conf_interval = best_model.predict(len(data_test), return_conf_int = True)
predicted = pd.Series(predicted, index =data_test.index)

y1 = np.exp(conf_interval[:,0])
y2 = np.exp(conf_interval[:,1])
np.exp(predicted).plot(style = '--', linewidth = 5)
np.exp(data_train).plot(style= 'o', alpha=0.2)
np.exp(data_test).plot(alpha = 0.8)
plt.fill_between(data_test.index, y1 , y2, color ='gray', alpha = 0.4)
```

# Prevision en utilisant Prophet

Prophet is a time series forecasting algorithm created by Facebook. Prophet is designed to simplify time series forecasting and is accessible to individuals without backgrounds in statistics, econometrics, or mathematics. Additionally, the algorithm is fast and accurate.


```python
import prophet
```

```python
prophet_model = prophet.Prophet ()
```

```python
df = pd.DataFrame ()
df ['ds'] = data_train.index
df ['y'] = data_train.values
```

```python
prophet_model.fit (df)
```

```python
df_test = pd.DataFrame ()
df_test ['ds'] = data_test.index
predictions = prophet_model.predict(df_test)
```
